```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data = requests.get(url).text
```

```python
dfs = pd.read_html(data, flavor='bs4')
```

```python
list1 = dfs[0].values.tolist() #convert datframe column to list
tdf = pd.DataFrame(list1[0],columns=['Column_Name'])
for x in range(19):
        dfa = pd.DataFrame (list1[x+1],columns=['Column_Name'])
        tdf = tdf.append(dfa, ignore_index=True,sort=False)
tdf.tail
```

```python
tdf['Column_Name'] = tdf['Column_Name'].values.astype('|S')
tdf.dtypes
```

```python
toronto_data = tdf
```

## Cleaning the data

```python
toronto_data['Postal_code'] =toronto_data["Column_Name"].str[0:3] #split()
toronto_data['Borough'] = toronto_data['Column_Name'].str[3:]
toronto_data['Borough'] = toronto_data['Borough'].values.astype(str)
toronto_data
```

```python
new = toronto_data['Borough'].str.split('(', n=1, expand=True)
toronto_data['Borough'] = new[0]
toronto_data['Neighborhood'] = new[1]
toronto_data
```

```python
toronto_data.drop(columns =["Column_Name"], inplace = True)
toronto_data
```

```python
toronto_data['Postal_code'] = toronto_data['Postal_code'].values.astype(str)
toronto_data['Neighborhood'] = toronto_data['Neighborhood'].values.astype(str)
```

```python
toronto_data
```

```python
toronto_data.shape
```

```python
toronto_data['Neighborhood'] = toronto_data['Neighborhood'].str.replace('/',',')
```

```python
toronto_data['Neighborhood'] = toronto_data['Neighborhood'].str.replace(')','')
```

```python
toronto_data
```

```python
indexNames = toronto_data[toronto_data['Borough'].str.contains('Not assigned')].index
# Delete these row indexes from dataFrame
indexNames
toronto_data.drop(indexNames , inplace=True)
toronto_data = toronto_data.reset_index(drop=True)
```

```python
toronto_data.head(10)
```

```python
toronto_data= toronto_data.sort_values('Postal_code')
```

```python
toronto_data.head(10)
```

```python
df2 = toronto_data.groupby(['Postal_code','Borough'], sort=False).agg(', '.join)
```

```python
df2
```

```python
toronto_data['Borough'] = toronto_data['Borough'].str.replace('EtobicokeNorthwest','Etobicoke')
toronto_data.Borough[toronto_data.Postal_code=='M7Y'] = 'East Toronto'
toronto_data.Borough[toronto_data.Postal_code=='M5W'] = 'Downtown Toronto'
toronto_data.Borough[toronto_data.Postal_code=='M4J'] = 'East York'
```

## Creating the dataframe with latitude and longitutde coordinates of each neighborhood

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
lat_lon.rename(columns={'Postal Code':'Postal_code'},inplace=True)
df3 = pd.merge(toronto_data,lat_lon,on='Postal_code')
df3.head(60)
```

## Working with only boroughs that contain the word Toronto 

```python
df4 = df3[df3['Borough'].str.contains('Toronto')]
df4
```

```python
df4 = df4.reset_index(drop=True)
```

```python
df4
```

## Visualizing data of the above dataframe

```python
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='blue',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

## Clustering Neighborhoods of the boroughs that contain the word Toronto

```python
from sklearn.cluster import KMeans

kclusters = 5

toronto_clustering = df4.drop(['Postal_code','Borough','Neighborhood'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4
```

```python
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
map_clusters
```
